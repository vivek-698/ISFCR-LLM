```python
from datasets import load_dataset

dataset = load_dataset("teknium/OpenHermes-2.5")
```

Showing column names and sample data:

```python
print("Column names:", dataset.column_names)
print("Sample data point:", dataset['train'][0])
```

Filter and select data:

```python
# Filter dialogues with long system utterances
filtered_dataset = dataset.filter(lambda example: len(example["conversations"]) > 50)
filtered_dataset
```

Now , Lets Pull a Base Model(Transfomer) and Try out a Task : https://huggingface.co/sgugger/finetuned-bert-mrpc


```python
from transformers import AutoModelForSequenceClassification,pipeline

model_name = "sgugger/finetuned-bert-mrpc"
model = AutoModelForSequenceClassification.from_pretrained(model_name)

classifier = pipeline("text-classification", model=model_name)
```

```python
classifier("I like this demo and i'll explore more tasks")
```

```python
model.push_to_hub("finetuned-bert-mrpc")
```
